### 2)Policies against objectives (aggregated over time)


```python
from ema_workbench import load_results
results2 = load_results('Week22_Open_exploration_1000_20.tar.gz')
```

```python
experiments, outcomes = results2
```

```python
experiments.columns
```

```python
x1=x2=x3=x4=x5=0
x_values = [x1,x2,x3,x4,x5]
numbers = [0,3,6,9,12]

for index in range(len(x_values)):
    c = numbers[index]
    x_values[index] = experiments.iloc[:, [(19 + c), (20 + c), (21 + c), (34 + c), (35 + c), (36 + c), 49]]

    
                         
    
```

```python
dikes = ['A.1', 'A.2','A.3', 'A.4', 'A.5']
for index2 in range(5):
    filtered_outcomes = {}
    for key in outcomes.keys():
        if key.startswith(dikes[index2]) or key.startswith('RfR') or key.startswith('Expected') :
            # If the condition is met, add the key-value pair to the filtered dictionary
            filtered_outcomes[key] = outcomes[key]
            # Calculate feature scores
    fs = feature_scoring.get_feature_scores_all(x_values[index2], filtered_outcomes)

    # Plot the heatmap
    sns.heatmap(fs, cmap="viridis", annot=True)
    plt.title(f'Feature scoring for dike {dikes[index2]}')
    plt.ylabel('Policies over time')
    plt.show()
```
